```python
import re

def extract_number(value):
    match = re.search(r'\d+', str(value))
    return int(match.group()) if match else None

dataset['Like'] = dataset['Like'].apply(extract_number)

```

```python
features = ['Age', 'Gender', 'VisitFrequency', 'Like',
            'yummy', 'cheap', 'healthy', 'fast', 'tasty', 'disgusting']

x = dataset[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
pca_result=pca.fit_transform(x_scaled)
```

```python
pca_df = pd.DataFrame(pca_result, columns=[f'PC{i+1}' for i in range(10)])
print(pca_df.head(10))
```

```python

loadings = pd.DataFrame(pca.components_.T[:,1:3],index=features,columns=['PC2', 'PC3'] )


plt.figure(figsize=(10, 8))

plt.scatter(pca_result[:, 1], pca_result[:, 2], color='lightgrey', alpha=0.6)

for i in range(len(loadings)):
    x = loadings.iloc[i, 0]  # PC2
    y = loadings.iloc[i, 1]  # PC3
    plt.arrow(0, 0, x, y, color='red', alpha=0.9, head_width=0.03)
    plt.text(x * 1.1, y * 1.1, loadings.index[i], color='red', ha='center', va='center', fontsize=9)


plt.xlabel('PC2')
plt.ylabel('PC3')
plt.title('PCA Biplot: PC2 vs PC3')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
psychographic = dataset[['yummy', 'cheap', 'healthy', 'fast', 'tasty', 'disgusting']]
scaled_psych = scaler.fit_transform(psychographic)
pca_psych = PCA(n_components=3)
pca_psych_result = pca_psych.fit_transform(scaled_psych)
print(pca_psych_result)
```

```python
demo_behav = dataset[['Age', 'Gender', 'VisitFrequency', 'Like']]
scaled_demo = scaler.fit_transform(demo_behav)
pca_demo = PCA(n_components=3)
pca_demo_result = pca_demo.fit_transform(scaled_demo)
print(pca_demo_result)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)


plt.figure(figsize=(8, 5))
plt.bar(range(1, 11), wcss, color='teal', edgecolor='black')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
clusters=kmeans.fit_predict(x_scaled)

dataset['Segment']=clusters
```

```python
segment_profile=dataset.groupby('Segment')[features].mean()
print(segment_profile)
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
X_pca=pca.fit_transform(x_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0],y=X_pca[:,1],hue=dataset['Segment'],palette='Set2')
plt.title('KMeans Customer Segments (PCA)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Segment')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.metrics.pairwise import pairwise_distances

linkage_matrix = linkage(x_scaled, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

num_clusters = 3
cluster_labels = fcluster(linkage_matrix, num_clusters, criterion='maxclust')

dataset['Cluster'] = cluster_labels

for i in range(1, num_clusters + 1):
    print(f"\n--- Cluster {i} profile ---")
    cluster_data = dataset[dataset['Cluster'] == i]
    print(cluster_data[features].describe(include='all'))
```
